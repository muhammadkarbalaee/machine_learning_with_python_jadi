# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel("funny numbers")
```

```python
plt.plot([[1, 2], [3, 4]])
plt.ylabel("funny numbers")
```

```python
plt.plot([[1, 2], [3, 4]], 'ro')
```

```python
plt.plot([[1, 2], [3, 4]], 'bo')
```

```python
plt.plot([[1, 2], [3, 4]], 'r+')
```

```python
t = np.arange(0, 5, 0.2)
t
```

```python
plt.plot(t, t ** 2, 'r--')
```

```python
plt.plot(t, t ** t, 'r-.')
```

```python
plt.plot(t, t ** t, 'r-.', t, t ** 1000, 'g-')
```
